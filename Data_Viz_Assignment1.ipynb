```python
#Let's Start by importing all the required library
import numpy as np
import matplotlib.pyplot as plt
#Generating random numbers
uniform_array = np.random.uniform(0, 1, 100)
plt.figure(figsize=(8, 5))
plt.boxplot(uniform_array, vert=True, patch_artist=True, boxprops=dict(facecolor='lightblue'))
#Adding label to x,y
plt.ylabel("Values ranging from 0 to 1")
plt.xlabel("Box Plot")
plt.grid(axis='y', linestyle='--', alpha=0.7)
#Adjustment
plt.tight_layout()
plt.figtext(0.5, -0.05, "Figure : Box Plot for Uniform Distribution", ha="center", va="center", fontsize=12)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
#Steps to calculate the mean and SD
mean = (1 + 100) / 2
std_dev = (100 - 1) / 6
gaussian_array = np.random.normal(mean, std_dev, 200)
plt.figure(figsize=(6, 8))
plt.boxplot(gaussian_array, vert=True, patch_artist=True, boxprops=dict(facecolor="lightgreen"))
plt.ylabel("Value ranging from 1 to 100")
plt.xlabel("Box Plot")
plt.figtext(0.5, -0.05, "Figure : Box Plot for Guassian Distribution", ha="center", va="center", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)
uniform_data = np.random.uniform(0, 1, 100)
gaussian_data = np.random.normal(50, 20, 200)

# This will help us to Create bins
bins = np.linspace(0, 1, 21)  # This one is for For uniform data
gaussian_bins = np.linspace(0, 100, 21)  # This one is for gaussian data
uniform_hist, _ = np.histogram(uniform_data, bins=bins)
gaussian_hist, _ = np.histogram(gaussian_data, bins=gaussian_bins)
#print(bins)
# Plot bar charts
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.bar(bins[:-1], uniform_hist, width=np.diff(bins), align='edge', alpha=0.7)
plt.title('Uniform Distribution')
plt.xlabel('Range Value')
plt.ylabel('Frequency')
plt.subplot(1, 2, 2)
plt.bar(gaussian_bins[:-1], gaussian_hist, width=np.diff(gaussian_bins), align='edge', alpha=0.7)
plt.title('Gaussian Distribution')
plt.xlabel('Range Value')
plt.ylabel('Frequency')
plt.tight_layout()
plt.figtext(0.5, -0.05, "Figure 2: Histogram created for both, the Uniform as well as the Guassian Distribution", ha="center", va="center", fontsize=12)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# RGenerating our random Data
np.random.seed(42)
uniform_data = np.random.uniform(0, 1, 100)
gaussian_data = np.random.normal(50, 20, 200)

# Write to binary file
with open('random_data.bin', 'wb') as f:
    np.save(f, uniform_data)
    np.save(f, gaussian_data)

# Read from binary file
with open('random_data.bin', 'rb') as f:
    read_uniform = np.load(f)
    read_gaussian = np.load(f)

#Now we will Plot Cumulative Distribution Functions
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sorted_uniform = np.sort(read_uniform)
plt.plot(sorted_uniform, np.arange(1, len(sorted_uniform) + 1) / len(sorted_uniform))
plt.title('Uniform Distribution CDF')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')

plt.subplot(1, 2, 2)
sorted_gaussian = np.sort(read_gaussian)
plt.plot(sorted_gaussian, np.arange(1, len(sorted_gaussian) + 1) / len(sorted_gaussian))
plt.title('Gaussian Distribution CDF')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.figtext(0.5, -0.05, "Figure 3: Cumulative Distribution Function for both the arrays", ha="center", va="center", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Again we will be creating array
uniform_array = np.random.uniform(0, 1, 100)
gaussian_array = np.random.normal(50.5, 16.5, 200)
# Uniform sampling in 2D
uniform_2d_x = np.random.uniform(0, 1, 5000)
uniform_2d_y = np.random.uniform(0, 1, 5000)
# Gaussian sampling in 2D
gaussian_2d_x = np.random.normal(0.5, 0.15, 5000)
gaussian_2d_y = np.random.normal(0.5, 0.15, 5000)
# Bin generation
uniform_hist, uniform_xedges, uniform_yedges = np.histogram2d(uniform_2d_x, uniform_2d_y, bins=100, range=[[0, 1], [0, 1]])
# Gaussian histogram
gaussian_hist, gaussian_xedges, gaussian_yedges = np.histogram2d(gaussian_2d_x, gaussian_2d_y, bins=100, range=[[0, 1], [0, 1]])
# Heatmaps for 2D histograms
plt.figure(figsize=(14, 6))
# Uniform heatmap
plt.subplot(1, 2, 1)
plt.imshow(uniform_hist.T, origin='lower', extent=[0, 1, 0, 1], aspect='auto', cmap='viridis')
plt.colorbar(label='Counts')
plt.title("Uniform Distribution")
plt.xlabel("X")
plt.ylabel("Y")
# Gaussian  heatmap
plt.subplot(1, 2, 2)
plt.imshow(gaussian_hist.T, origin='lower', extent=[0, 1, 0, 1], aspect='auto', cmap='viridis')
plt.colorbar(label='Counts')
plt.title("Gaussian Distribution")
plt.xlabel("X")
plt.ylabel("Y")
plt.tight_layout()
plt.figtext(0.5, -0.05, "Figure 4b: 2D Array of size 100x100", ha="center", va="center", fontsize=12)
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)
uniform_array = np.random.uniform(0, 1, 100)

# Generate Gaussian distribution array between 1 and 100 and we also calculate the mean
gaussian_array = np.random.normal(loc=50, scale=25, size=200)

# Then create figure and subplots
plt.figure(figsize=(12, 5))

# Uniform distribution contour plot
plt.subplot(121)
x = np.random.uniform(0, 1, 100)
y = np.random.uniform(0, 1, 100)
plt.tricontourf(x, y, uniform_array, levels=10, cmap='viridis')
plt.colorbar(label='Uniform Values')
plt.title('Uniform Distribution Contour')
plt.xlabel('X')
plt.ylabel('Y')

# Gaussian distribution contour plot
plt.subplot(122)
x = np.random.uniform(0, 100, 200)
y = np.random.uniform(0, 100, 200)
plt.tricontourf(x, y, gaussian_array, levels=10, cmap='plasma')
plt.colorbar(label='Gaussian Values')
plt.title('Gaussian Distribution Contour')
plt.xlabel('X')
plt.ylabel('Y')
plt.figtext(0.5, -0.05, "Figure 4c:  Uniform and Gaussian sampled arrays as contour plots", ha="center", va="center", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load and clean the dataset
file_path = r'/content/NOAA-Temperatures.csv'
data = pd.read_csv(file_path, skiprows=4, names=["Year", "Degrees_F"], usecols=[0, 1])

# This is kind of data pre processing.
# Drop rows with missing or invalid data
data = data.dropna()
data = data[data["Year"].str.isdigit()]

# Convert columns to appropriate data types.
data["Year"] = data["Year"].astype(int)
data["Degrees_F"] = pd.to_numeric(data["Degrees_F"], errors="coerce")

# Define colors based on positive or negative anomalies
data["Color"] = data["Degrees_F"].apply(lambda x: "red" if x > 0 else "blue")

# Plot the bar chart
plt.figure(figsize=(14, 7))
plt.bar(data["Year"], data["Degrees_F"], color=data["Color"], width=1.0)

# Add labels and title
plt.xlabel("Year", fontsize=14)
plt.ylabel("Degrees F +/- From Average", fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
# Show plot
plt.figtext(0.5, -0.05, "Figure 2.1:  Change in Temperature over a certain Time Frame", ha="center", va="center", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import openpyxl

# Load the dataset
file_path = r'/content/Breakfast-Cereals.xls'
data = pd.read_excel(file_path, sheet_name='cereals')
# Filter the dataset for the cereals. In this case we only want 3.
selected_cereals = data[data['Cereal'].isin(['Froot Loops', 'Honey Nut Cheerios', 'Quaker Oatmeal'])]
cereal_names = selected_cereals['Cereal']

# Select nutritional statistics
nutritional_stats = ['Calories', 'Protein', 'Fat', 'Sodium', 'Fiber', 'Carbohydrates', 'Sugars', 'Potassium']
stats = selected_cereals[nutritional_stats].values

# Compute angles for the radar chart
num_vars = len(nutritional_stats)
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]
fig, ax = plt.subplots(figsize=(7,7), subplot_kw=dict(polar=True))

# Plot each cereal, we will draw the chart
for i, cereal in enumerate(stats):
    values = cereal.tolist()
    values += values[:1]  # Close the circle
    ax.plot(angles, values, label=cereal_names.iloc[i])
    ax.fill(angles, values, alpha=0.25)

# Add labels and title
ax.set_yticks([])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(nutritional_stats, fontsize=10)
#ax.set_title('', fontsize=14, pad=20)
ax.legend(loc='upper right', bbox_to_anchor=(1.32, 1.1), fontsize=10)
# Display the chart
plt.figtext(0.5, -0.05, "Figure 2.2:  Nutritional Comparison of Froot Loops, Honey Nut Cheerios, and Quaker Oatmeal", ha="center", va="center", fontsize=12)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import parallel_coordinates

# Lets load the dataset
file_path = "/content/drinks.csv"  # Update this path if needed
df = pd.read_csv(file_path)

# Normalize numerical data for visualization, basically we are organising our data for better representation.
df_numeric = df.drop(columns=["country"])
df_normalized = (df_numeric - df_numeric.min()) / (df_numeric.max() - df_numeric.min())
df_normalized["country"] = df["country"]  # We here add the country names back for labeling

# Plot parallel coordinates, coloring by country
plt.figure(figsize=(12, 6))
parallel_coordinates(df_normalized, class_column="country", colormap=plt.cm.viridis, alpha=0.7)
plt.title("Parallel Coordinates Plot of Alcohol Consumption")
plt.xlabel("Features")
plt.ylabel("Normalized Values")
plt.xticks(rotation=20)
plt.legend([], [], frameon=False)
plt.grid(True)
plt.show()
```
